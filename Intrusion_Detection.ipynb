```python
cat = ['protocol_type','service','flag']
Encoder = LabelEncoder()
for i in cat:
    df[i] = Encoder.fit_transform(df[i])
df
```

# **Feature Selection**

```python
from sklearn.ensemble import ExtraTreesClassifier
x = df.drop(columns='class')
y = df['class']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
model = ExtraTreesClassifier(n_estimators=100)
model.fit(x_train,y_train)
predict = model.predict(x_test)
predict
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,predict)*100
score
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary to store the metrics
model_metrics = {
    'Model': ['Random Forest', 'Logistic Regression', 'KNN', 'Decision Tree', 'AdaBoost', 'SVM', 'XGBoost','Extra Tree'],
    'Accuracy': [Rascore, Lascore, Kascore, Dascore, Aascore, Sscore, Xscore],
    'Precision': [Rpscore, Lpscore, Kpscore, Dpscore, Apscore, Spscore, Xpscore],
    'Recall': [Rrecall, Lrecall, Krecall, Drecall, Arecall, Srecall, Xrecall],
    'F1-Score': [Rfscore, Lfscore, Kfscore, Dfscore, Afscore, Sfscore, Xfscore]
}

# Create a Pandas DataFrame
metrics_df = pd.DataFrame(model_metrics)
# Set the width of the bars
bar_width = 0.2

# Set the positions of the bars on the x-axis
r1 = np.arange(len(metrics_df))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Create the bar graph
plt.bar(r1, metrics_df['Accuracy'], color='b', width=bar_width, edgecolor='grey', label='Accuracy')
plt.bar(r2, metrics_df['Precision'], color='g', width=bar_width, edgecolor='grey', label='Precision')
plt.bar(r3, metrics_df['Recall'], color='r', width=bar_width, edgecolor='grey', label='Recall')
plt.bar(r4, metrics_df['F1-Score'], color='y', width=bar_width, edgecolor='grey', label='F1-Score')

# Add labels, title, and legend
plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Model Performance Comparison')
plt.xticks([r + bar_width for r in range(len(metrics_df))], metrics_df['Model'], rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
